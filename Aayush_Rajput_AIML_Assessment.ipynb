# Some preprocessing on test dataset

```python
# Load the testing dataset
testing_file_path = "/kaggle/input/train-and-test/AP_Moller_Testing Data.xlsx"
testing_df = pd.read_excel(testing_file_path)

# Display the first few rows of the testing dataset
print(testing_df.head())
```

```python
# Separate features (X) and target variable (y) for training set
X_train = df.drop(['Month of Sourcing', 'Sourcing Cost'], axis=1)
y_train = df['Sourcing Cost']

# Separate features (X) and target variable (y) for testing set
X_test = testing_df.drop(['Month of Sourcing', 'Sourcing Cost'], axis=1)
y_test = testing_df['Sourcing Cost']

# Display the shapes of training and testing sets
print("Training set shape:", X_train.shape, y_train.shape)
print("Testing set shape:", X_test.shape, y_test.shape)
```

```python
# Check the data types of features in the training dataset
print(X_train.dtypes)

# Check for any non-numeric values in the training dataset
print(X_train[X_train.applymap(lambda x: isinstance(x, str))].stack().unique())
```

# Linear Regression

```python
from sklearn.metrics import mean_squared_error, r2_score  # Import necessary libraries
from sklearn.linear_model import LinearRegression

# Apply one-hot encoding to categorical features in the training dataset
X_train_encoded = pd.get_dummies(X_train)

# Initialize and train the Linear Regression model
lr_model = LinearRegression()
lr_model.fit(X_train_encoded, y_train)

# Make predictions on the testing set
X_test_encoded = pd.get_dummies(X_test)
y_pred_lr = lr_model.predict(X_test_encoded)

# Evaluate the model using RMSE
rmse_lr = mean_squared_error(y_test, y_pred_lr, squared=False)
print("RMSE for Linear Regression:", rmse_lr)

# Calculate R-squared score
r2_lr = r2_score(y_test, y_pred_lr)
print("R-squared score for Linear Regression:", r2_lr)
```

# Decision Tree Model

```python
from sklearn.tree import DecisionTreeRegressor

# Initialize and train the Decision Tree model
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train_encoded, y_train)

# Make predictions on the testing set
y_pred_dt = dt_model.predict(X_test_encoded)

# Evaluate the model using RMSE, MSE, and R-squared score
rmse_dt = mean_squared_error(y_test, y_pred_dt, squared=False)
r2_dt = r2_score(y_test, y_pred_dt)

# Display the performance metrics
print("Decision Tree Model:")
print("RMSE:", rmse_dt)
print("R-squared:", r2_dt)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize and train the Random Forest model
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train_encoded, y_train)

# Make predictions on the testing set
y_pred_rf = rf_model.predict(X_test_encoded)

# Evaluate the model using RMSE, MSE, and R-squared score
rmse_rf = mean_squared_error(y_test, y_pred_rf, squared=False)
r2_rf = r2_score(y_test, y_pred_rf)

# Display the performance metrics
print("Random Forest Model:")
print("RMSE:", rmse_rf)
print("R-squared:", r2_rf)
```

# Gradient Boosting Regression

```python
from sklearn.ensemble import GradientBoostingRegressor

# Initialize and train the Gradient Boosting model
gb_model = GradientBoostingRegressor(random_state=42)
gb_model.fit(X_train_encoded, y_train)

# Make predictions on the testing set
y_pred_gb = gb_model.predict(X_test_encoded)

# Evaluate the model using RMSE, MSE, and R-squared score
rmse_gb = mean_squared_error(y_test, y_pred_gb, squared=False)
r2_gb = r2_score(y_test, y_pred_gb)

# Display the performance metrics
print("Gradient Boosting Model:")
print("RMSE:", rmse_gb)
print("R-squared:", r2_gb)
```

# Optimizing Decision Tree

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor

# Define the hyperparameters grid for Decision Tree
param_grid_dt = {
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize the Decision Tree regressor
dt_model = DecisionTreeRegressor(random_state=42)

# Initialize GridSearchCV
grid_search_dt = GridSearchCV(dt_model, param_grid_dt, cv=5, scoring='neg_mean_squared_error', verbose=1, n_jobs=-1)

# Perform GridSearchCV
grid_search_dt.fit(X_train_encoded, y_train)

# Get the best hyperparameters and best estimator
best_params_dt = grid_search_dt.best_params_
best_estimator_dt = grid_search_dt.best_estimator_

# Print the best hyperparameters
print("Best Hyperparameters for Decision Tree:", best_params_dt)

# Make predictions on the testing set using the best estimator
y_pred_dt_tuned = best_estimator_dt.predict(X_test_encoded)

# Evaluate the model using RMSE, MSE, and R-squared score
rmse_dt_tuned = mean_squared_error(y_test, y_pred_dt_tuned, squared=False)
r2_dt_tuned = r2_score(y_test, y_pred_dt_tuned)

# Display the performance metrics
print("Performance Metrics for Decision Tree after Hyperparameter Tuning:")
print("RMSE:", rmse_dt_tuned)
print("R-squared:", r2_dt_tuned)
```

# LSTM

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler

# Encode categorical features using one-hot encoding
encoder = OneHotEncoder(sparse=False, drop='first')
X_train_encoded = encoder.fit_transform(X_train.select_dtypes(include=['object']))

# Concatenate one-hot encoded features with numerical features
X_train_concatenated = np.concatenate((X_train_encoded, X_train.select_dtypes(exclude=['object'])), axis=1)

# Scale the concatenated features using MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(X_train_concatenated)

# Reshape the data for LSTM input (samples, time steps, features)
n_features = scaled_data.shape[1]
X_train_reshaped = scaled_data.reshape((scaled_data.shape[0], 1, n_features))

# Define and compile the LSTM model
model_lstm = Sequential()
model_lstm.add(LSTM(50, activation='relu', input_shape=(1, n_features)))
model_lstm.add(Dense(1))
model_lstm.compile(optimizer='adam', loss='mse')

# Fit the LSTM model
model_lstm.fit(X_train_reshaped, y_train, epochs=25, batch_size=32, verbose=1)
```

# Result of LSTM

```python
# Preprocess the testing dataset (similar to training dataset preprocessing)
X_test_encoded = encoder.transform(X_test.select_dtypes(include=['object']))
X_test_concatenated = np.concatenate((X_test_encoded, X_test.select_dtypes(exclude=['object'])), axis=1)
scaled_test_data = scaler.transform(X_test_concatenated)
X_test_reshaped = scaled_test_data.reshape((scaled_test_data.shape[0], 1, scaled_test_data.shape[1]))

# Use the trained LSTM model to make predictions on the testing data
y_pred_lstm = model_lstm.predict(X_test_reshaped)

# Calculate performance metrics
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

mse_lstm = mean_squared_error(y_test, y_pred_lstm)
rmse_lstm = np.sqrt(mse_lstm)
r2_lstm = r2_score(y_test, y_pred_lstm)

print("Performance Metrics for LSTM on Testing Dataset:")
print("RMSE:", rmse_lstm)
print("MSE:", mse_lstm)
print("R-squared:", r2_lstm)
```

# GRU

```python
from keras.models import Sequential
from keras.layers import GRU, Dense
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# One-hot encode categorical features
X_train_encoded = pd.get_dummies(X_train)

# Scale the data using MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(X_train_encoded)

# Reshape the data for GRU input (samples, time steps, features)
n_features = X_train_encoded.shape[1]
X_train_reshaped = scaled_data.reshape((scaled_data.shape[0], 1, n_features))


# Define the GRU model
model_gru = Sequential()
model_gru.add(GRU(50, activation='relu', input_shape=(1, n_features)))
model_gru.add(Dense(1))
model_gru.compile(optimizer='adam', loss='mse')

# Fit the GRU model
model_gru.fit(X_train_reshaped, y_train, epochs=100, batch_size=32, verbose=1)
```

# Result of GRU

```python
from sklearn.preprocessing import OneHotEncoder

# Concatenate training and testing datasets
concatenated_df = pd.concat([X_train, X_test])

# Perform one-hot encoding
encoder = OneHotEncoder()
encoded_data = encoder.fit_transform(concatenated_df)

# Convert sparse matrix to dense array
encoded_data_dense = encoded_data.toarray()

# Split back into training and testing datasets
X_train_encoded = encoded_data_dense[:len(X_train)]
X_test_encoded = encoded_data_dense[len(X_train):]

# Scale the data using MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_train_data = scaler.fit_transform(X_train_encoded)
scaled_test_data = scaler.transform(X_test_encoded)

# Reshape the data for GRU input (samples, time steps, features)
n_features = X_train_encoded.shape[1]
X_train_reshaped = scaled_train_data.reshape((scaled_train_data.shape[0], 1, n_features))
X_test_reshaped = scaled_test_data.reshape((scaled_test_data.shape[0], 1, n_features))

from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Use the trained GRU model to make predictions on the testing data
y_pred_gru = model_gru.predict(X_test_reshaped)

# Calculate performance metrics
mse_gru = mean_squared_error(y_test, y_pred_gru)
rmse_gru = np.sqrt(mse_gru)
r2_gru = r2_score(y_test, y_pred_gru)

print("Performance Metrics for GRU on Testing Dataset:")
print("RMSE:", rmse_gru)
print("MSE:", mse_gru)
print("R-squared:", r2_gru)
```
